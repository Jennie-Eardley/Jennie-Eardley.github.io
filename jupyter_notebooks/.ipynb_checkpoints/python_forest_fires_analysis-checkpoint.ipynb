Forest Fires Analysis

```python
# loading in libraries for analysis and visualisation
import pandas as pd
import pandas_profiling
import numpy as np
import seaborn as sns
```

```python
# loading in data 
fires = pd.read_csv("forestfires.csv")
```

```python
# looking at summary statistics
fires.describe()
```

```python
# looking at a summary report - indicates missing values and contains some useful visualisations
pandas_profiling.ProfileReport(fires)
```

```python
# replacing area NaN values with 0
fires["area"].fillna(0,inplace = True)
```

```python
# checking to ensure that they have been correctly replaced
fires.isnull().sum()
```

```python
# removing x and y variables 
fires.drop(["X", "Y"], axis = 1, inplace = True)
```

```python
# sorting data by area 
fires.sort_values("area", ascending = False, inplace = True)
```

```python
# viewing the first five lines to check that data is sorted by area
fires.head()
```

```python
# creating a filtered dataframe with values included only where area is known 
filtered_fires = fires[fires["area"] > 0]
```

```python
# creating a new boolean variable which indicates whether or not it has rained 
filtered_fires.loc[filtered_fires['rain'] > 0, 'is_rain'] = 'True'
filtered_fires.loc[filtered_fires['rain'] == 0, 'is_rain'] = 'False'
```

```python
# looking at the mean area of fire by month 
filtered_fires.groupby('month')['area'].mean()
```

```python
# plotting temperature against area 
sns.regplot(x = 'temp', y = 'area', data = filtered_fires)
```
